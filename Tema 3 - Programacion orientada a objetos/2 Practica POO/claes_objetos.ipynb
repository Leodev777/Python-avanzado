```python
class Persona:
    def __init__(self, nombre, edad, profesion):
        self.nombre = nombre
        self.edad = edad
        self.profesion = profesion

person1 = Persona("Pepe", 58,"Abogado")
person2 = Persona("Maria", 21,"Estudiante")

print(person1.nombre, person1.edad, person1.profesion)
print(person2.nombre, person2.edad, person2.profesion)
```

```python
class calculadora:
    def sumar(self,a,b):
        return a + b
    def resta(self,a,b):
        return a - b
    def multi(self,a,b):
        return a * b
    def div(self,a,b):
        return a / b

calculadora = calculadora() # Creo la instancia
print(calculadora.sumar(1,3))
print(calculadora.resta(7,3))
print(calculadora.multi(7,3))
print(calculadora.div(21,3))
print(calculadora.sumar(300,3))
```

```python
class Libro: # Clase libro con los 3 atributos
    def __init__(self, titulo, autor, anio_public):
        self.titulo = titulo
        self.autor = autor
        self.anio_public = anio_public

# Objetos
lib1 = Libro("Python", "Ramon diaz", "2020")
lib2 = Libro("C++", "LocoLuis", "2023")
lib3 = Libro("Solidity", "Leonardo", "2056")

print(lib1.titulo, lib1.autor, lib1.anio_public)
print(lib2.titulo, lib2.autor, lib2.anio_public)
print(lib3.titulo, lib3.autor, lib3.anio_public)
```

```python
class rectangulo:
    def __init__(self, longitud, ancho):
        self.longitud = longitud
        self.ancho = ancho
    
    def calcular_area(self):
        area = self.longitud * self.ancho
        return area
    
    def cal_perimetro(self):
        return 2 * (self.longitud + self.ancho)
    

rec1 = rectangulo(5,3)
print("Area", rec1.calcular_area())
print("Perimetro", rec1.cal_perimetro())
```

```python

import random

class Dado:
    def lanzar_dado(self):
        lanzamiento = random.randint(1,6)
        return lanzamiento

dado = Dado()
dado2 = Dado()
resultado = dado.lanzar_dado()
resultado2 = dado2.lanzar_dado()
print("Resultado del dado: ", resultado)
print("Resultado del dado2: ", resultado2)


```

```python
class Coche:
    def __init__(self, marca, modelo, anio):
        self.marca = marca
        self.modelo = modelo
        self.anio = anio
        self.encencido = False
    
    def encender(self):
        self.encencido = True

    def apagar(self):
        self.encencido = False


mi_coche = Coche("Toyota", "Corolla", 2015)
mi_coche.encender()
print("El coche esta encendido:", mi_coche.encencido)
mi_coche.apagar()
print("El coche esta encendido:", mi_coche.encencido)
```
