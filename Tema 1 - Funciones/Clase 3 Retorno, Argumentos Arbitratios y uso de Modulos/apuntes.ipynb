Valores de Retorno:
Si queremos añadir una entrada que sea opcional, como por ejemplo un segundo nombre! podemos decarar la varible de la funcion de la siguiente forma recibir_nombre_formateado(nombre, segundo nombre = "", apellido)

```python
def recibir_nombre_formateado(nombre, apellido):
    """Devuelve el nombre completo bien formateado."""
    nombre_completo = nombre + " " + apellido
    return nombre_completo.title()

musico = recibir_nombre_formateado('jimi', 'hendrix')
print(musico) # Salida: Jimi Hendrix
```

Valores de retorno, Diccionarios:

```python
def construir_persona(nombre, apellido):
    """Devuelve un diccionario con informacion de la persona."""
    persona = {'nombre': nombre, 'apellido': apellido}
    return persona
usuario = construir_persona("Juan", "Perez")
print(usuario)
```

```python
def construir_persona(nombre, apellido, edad = ""):
    """Devuelve un diccionario con informacion de la persona."""
    persona = {'nombre': nombre, 'apellido': apellido}
    if edad:
        persona['edad'] = edad
    return persona
usuario = construir_persona("Juan", "Perez", 555)
print(usuario)
```

Uso de funciones en bucles:

```python
def recibir_nombre_formateado(nombre, apellido):
    """Devuelve el nombre completo bien formateado."""
    nombre_completo = nombre + '' + apellido
    return nombre_completo.title()
# Estro es el bucle infinito
while True:
    print("\nDime tu nombre completo") 
    nombre = input("Nombre: ")
    apellido = input("apellido: ")
    
    nombre_completo = recibir_nombre_formateado(nombre, apellido)
    print("Hola, " + nombre_completo + "!")
```

Listas en funciones:

```python
def saludar_usuario(lista_nombres):
    """Saludar a cada usuario de la lista."""
    for nombre in lista_nombres:
        mensaje = "Hola, " + nombre.title() + "."
        print(mensaje)

usuarios = ["Juan", "Maria", "Luisilloelpillo"]
saludar_usuario(usuarios)
    
```

Listas en funciones, Modificar una lista:

```python
# Lista de diseños que deben ser impresos en 3D
encargos = ["Telefono", "Robot", "Asesinadordefuego"]
finalizados = []
# Simulamos que se eimprimen los diseños hasta que no queda ninguno, mo vemos los dieseños impresos a la lista de finalizados
while encargos:
    diseño_actual = encargos.pop()
    # Simulamos la creacion del sieño en 3D
    print("iImprimiendo el diseño: " + diseño_actual)
    finalizados.append(diseño_actual)
# Mostrar diseños eimpresos
print("\nLos siguientes diseños han sido impresos:")
for modelo_finalizado in finalizados:
    print(modelo_finalizado)
```
