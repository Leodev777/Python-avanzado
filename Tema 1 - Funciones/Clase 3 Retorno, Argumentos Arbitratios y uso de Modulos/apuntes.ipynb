Valores de Retorno:
Si queremos a√±adir una entrada que sea opcional, como por ejemplo un segundo nombre! podemos decarar la varible de la funcion de la siguiente forma recibir_nombre_formateado(nombre, segundo nombre = "", apellido)

```python
def recibir_nombre_formateado(nombre, apellido):
    """Devuelve el nombre completo bien formateado."""
    nombre_completo = nombre + " " + apellido
    return nombre_completo.title()

musico = recibir_nombre_formateado('jimi', 'hendrix')
print(musico) # Salida: Jimi Hendrix
```

Valores de retorno, Diccionarios:

```python
def construir_persona(nombre, apellido):
    """Devuelve un diccionario con informacion de la persona."""
    persona = {'nombre': nombre, 'apellido': apellido}
    return persona
usuario = construir_persona("Juan", "Perez")
print(usuario)
```

```python
def construir_persona(nombre, apellido, edad = ""):
    """Devuelve un diccionario con informacion de la persona."""
    persona = {'nombre': nombre, 'apellido': apellido}
    if edad:
        persona['edad'] = edad
    return persona
usuario = construir_persona("Juan", "Perez", 555)
print(usuario)
```
