```python
def construir_perfil (nombre, apellido, **informacion_user): # ** indica que puede recibir un numero variable de argumentos con clave-valor
    """Construir un diccionario conteniendo todo lo que sabermos del usuario"""
    perfil = {} # Creamos un diccionario vacio
    perfil[nombre] = nombre # Agregamos el nombre al diccionario
    perfil[apellido] = apellido # Agregamos el apellido al diccionario

    print(informacion_user) # Imprimimos el diccionario con la infomraccion
    print(type(informacion_user)) # Imprimimos el tipo de dato del diccionario

    for clave, valor in informacion_user.items(): # Recorremos el diccionario con un ciclo for, items() devuelve una lista de tuplas con clave-valor
        perfil[clave] = valor # Agregamos cada clave-valor al diccionario perfil
        return perfil # Devolvemos el diccionario perfil

    perfil_usuario = construir_perfil('juan', 'perez', edad=30, ciudad='madrid', profesion='ingeniero') # Llamamos a la funcion con argumentos clave-valor
    print(perfil_usuario) # Imprimimos el diccionario perfil_usuario
```
